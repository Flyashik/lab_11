```python
import os
import numpy as np
import cv2 as cv
import pickle
from torchvision.models import resnet50
import torch
import clip
from PIL import Image
```

```python
device = "cuda" if torch.cuda.is_available() else "cpu"
model, preprocess = clip.load("ViT-B/32", device=device)
```

```python
image = preprocess(Image.open("VOCdevkit\\VOC2012\\JPEGImages\\2007_000027.jpg")).unsqueeze(0).to(device)

with torch.no_grad():
    image_features = model.encode_image(image)
```

```python
import os
testDir = "VOCdevkit\\VOC2012\\JPEGImages"
imageArray = os.listdir(testDir) 
cwd = os.getcwd()
```

```python
id = 0
table = []
for imgName in imageArray: 
    path = cwd + "\\" + testDir + "\\" + imgName
    image = preprocess(Image.open(path)).unsqueeze(0).to(device)   
    with torch.no_grad():
        vec = model.encode_image(image)[0].numpy()
    if len(vec) == 0:
        continue
    dicte = {"id": id, "path":path, "vector":vec}
    id+=1
    table.append(dicte)
```

```python
import pandas as pd
data = pd.DataFrame(table)
```

```python
data['vector'].values
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
model = NearestNeighbors(n_neighbors=5,
                            metric='cosine',
                            algorithm='brute',
                            n_jobs=-1)

model.fit(np.stack(data['vector']))
```

```python
vec = preprocess(Image.open("VOCdevkit\\VOC2012\\JPEGImages\\2007_001439.jpg")).unsqueeze(0).to(device)
with torch.no_grad():
    vec = model.encode_image(image)[0].numpy()
```
